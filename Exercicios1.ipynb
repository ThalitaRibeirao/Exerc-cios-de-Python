1.
Desenvolva um programa que calcula se um dado ponto (x, y) está dentro de uma circunferência de centro (a, b) e raio r. Considere que pontos na linha da circunferência contam como dentro da circunferência.

````
p_x = float(input("Dê o valor de x do ponto: "))  
p_y = float(input("Dê o valor de y do ponto: "))  
circulo_x = float(input("Dê o valor de x da circunferência: "))   
circulo_y =  float(input ("Dê o valor de y da circunferência: "))  
raio = float(input("Dê o raio da circuferência: "))   
distancia_x = float((p_x - circulo_x)**2)  
distancia_y = float((p_y - circulo_y)**2)    
distancia_total = float((distancia_x + distancia_y)**(1/2))   
if distancia_total <= raio:
  print ("está dentro da circuferência")
else: 
  print ("está fora da circuferência")
````
